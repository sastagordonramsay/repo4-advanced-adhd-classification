```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from nilearn import datasets
```

```python
adhd = datasets.fetch_adhd(n_subjects=80)

print("Scans downloaded:", len(adhd.func))
```

```python
pheno = adhd.phenotypic.copy()

participants = pd.DataFrame({
    "subject_id": pheno["Subject"],
    "age": pheno["age"],
    "sex": pheno["sex"],
    "site": pheno["site"],
    "label": pheno["adhd"]
})

participants.head()
```

```python
print(participants["label"].value_counts())
```

```python
sns.countplot(x="label", data=participants)
plt.title("ADHD vs Control Counts")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(
    data=participants,
    x="age",
    hue="label",
    bins=12,
    kde=True
)
plt.title("Age Distribution by Group")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=participants, x="site", hue="label")
plt.xticks(rotation=45)
plt.title("Site Distribution by Group")
plt.show()
```

```python
os.makedirs("data/processed", exist_ok=True)

participants.to_csv(
    "data/processed/participants_expanded.csv",
    index=False
)

print("Saved metadata")
```
